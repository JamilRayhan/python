```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = np.arange(-5, 5, 0.1)
y1 = 0.25 * np.cos(np.pi * (n - 3)) + 0.25 * np.cos(np.pi * n) 
plt.subplot(3,1,1)
plt.plot(n, y1)
plt.title("y1(n) = 1/4 * cos((n-3)) + 1/4 * cos(n)")
plt.xlabel("n")
plt.ylabel("y1(n)")
plt.grid(True)
plt.show()

y2 = 2 * np.sin(np.pi * (n + 3)) + 2 * np.sin(np.pi * (n + 1)) 

plt.subplot(3,1,2)
plt.plot(n, y2)
plt.title("y2(n) = 2 * sin((n+3)) + 2 * sin((n+1))")
plt.xlabel("n")
plt.ylabel("y2(n)")
plt.grid(True)
plt.show()

y=np.add(y1,y2)

plt.subplot(3,1,3)
plt.plot(n, y)
plt.title("y(n) = add(y1,y2)")
plt.xlabel("n")
plt.ylabel("y(n)")
plt.grid(True)
plt.show()

if np.any(np.diff(y) != 0):
    print("Yes")
else:
    print("No")
```
